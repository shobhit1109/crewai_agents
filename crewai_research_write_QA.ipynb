```python
import dotenv
import os
dotenv.load_dotenv()
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
```

```python
from crewai import Agent, Task, Crew, Process

# Define Agents
researcher = Agent(
    role="Domain Researcher",
    goal="Collect accurate, relevant points with clarity about the topic",
    backstory="You summarize like a consultant",
    verbose=True
)

writer = Agent(
    role="Explainer",
    goal= "Convert research into step by step guide for beginners",
    backstory = "you explain things like a mentor with examples and structure",
    verbose=True
)

reviewer = Agent(
    role="Reviewer",
    goal="Check accuracy , remove fluff, ensure beginner friendly and clarity",
    backstory="You are strict about accuracy and clarity, you are a mentor who ensures the content is beginner friendly",
    verbose=True
)

# Define Tasks
research_task = Task(
    name="Research Task",
    description="Conduct research and create a structured outline to teach Crew AI to beginners",
    expected_output="8 section headings with 3 bullet points each",
    agent=researcher
)

write_task = Task(
    name="Write Task",
    description="Write a beginner guide based on the research outline provided by the Researcher. Use simple language and provide one example at the end",
    expected_output="Step by step guide with one example for each section",
    agent=writer
)

review_task = Task(
    name="Review Task",
    description="Review the written guide for accuracy, clarity, and beginner-friendliness.",
    expected_output="Final version of the guide with all errors corrected",
    agent=reviewer
)
```

```python
# Create a Crew and add tasks
crew = Crew(
    name="Content Creation Crew",
    agents=[researcher, writer,reviewer],
    description="A crew of agents working together to create content about Crew AI.",
    tasks=[research_task, write_task,review_task]
    )
```

```python
#Run
result = crew.kickoff()
print(result)
```
